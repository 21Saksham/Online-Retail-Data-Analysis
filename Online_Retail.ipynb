```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Online Retail.csv')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data.head()
```

```python
data.shape
```

```python
data = data.dropna()
```

```python
data.isna().sum()
```

## Removing Invalid data containing Unit Price Less than equal to 0

```python
data = data.drop(data[data['UnitPrice'] <= 0].index)
```

## Removing Invalid data containing Quantity Less than to 0

```python
data = data.drop(data[data['Quantity'] < 1].index)
```


## Removing Invalid data containing Unspecified Country Name

```python
data = data.drop(data[data['Country'] == 'Unspecified'].index)
```

```python
data.shape
```

## Remove Duplicates

```python
duplicate = data[data.duplicated(keep=False)].sort_values(by='Description')
duplicate.shape
```

```python
data = data.drop_duplicates()
data.shape
```

```python
data = data.drop(data[data['Description']=='Manual'].index)
data.shape
```

## Datatype Correction

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data['CustomerID'] = data['CustomerID'].astype(int)
data['InvoiceNo'] = data['InvoiceNo'].astype(int)
```

```python
data.info()
```

```python
data = data.reset_index()
```

## Revenue Generated by Products

```python
data['Revenue'] = data['UnitPrice']*data['Quantity']
```

```python
# Top 5 revenue generating products, there unit Price, Quantity & revenue generated
df = data.groupby(['Description','UnitPrice','Quantity'])['Revenue'].sum().sort_values(ascending = False)
df.head()
```

```python
# Bottom 5 revenue generating products, there unit Price, Quantity & revenue generated
df = data.groupby(['Description','UnitPrice','Quantity'])['Revenue'].sum().sort_values()
df.head()
```

## Revenue Generated By Countries

```python
# Top Revenue Generating Country
df = data.groupby(['Country'])['Revenue'].sum().sort_values(ascending = False).head()
df.head()
```

```python
# Top 10 revenue generating Products and Revenue Generated in United Kingdom
United_Kingdom = data[data['Country'] == 'United Kingdom']
df = United_Kingdom.groupby(['Description'])['Revenue'].sum().sort_values(ascending = False)
df.head(10)
```

```python
# Bottom Revenue Generating Country
df = data.groupby(['Country'])['Revenue'].sum().sort_values()
df.head()
```

```python
# Bottom 10 revenue generating Products and Revenue Generated in Saudi Arabia
Saudi_Arabia = data[data['Country'] == 'Saudi Arabia']
df = Saudi_Arabia.groupby(['Description'])['Revenue'].sum().sort_values()
df.head(10)
```
